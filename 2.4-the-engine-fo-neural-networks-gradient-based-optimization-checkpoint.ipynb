## Gradient Descent

非线性转换：因为在经过非线性转换后，数据可能会变得线性可分，而且可以加的层数变多。


关于神经网络的公式：
$\sigma(w^3\sigma(w^2\sigma(w^1x+b^1)+b^2)+b^3)$

<img src='./images/神经网络计算图01.png'>

这里w^2是w2，下划线1个=向量，下划线2个=矩阵，σ=非线性转换，b=截距项

梯度训练的方法：
+ SGD : 每次只丢一条数据去训练，这样就需要更新样本量集的次数
+ minibatch-SGD : 小量梯度下降算法
+ batch-SGD : 全量梯度下降算法

牛顿法的理解：
1 牛顿法主要的意思是给定一条曲线$f(x)$,,我们希望找到与x轴相交的点$x_{zero}$，现在已知曲线上的点(x0,y0)，然后我们通过$y-y_0=f^{\prime}(x_0)(x-x_0)$函数找到与x轴相交的点x_1,再将x_1带入f(x),得到$x_0 * \frac{-y_0}{f^{\prime}(x_0)}=x$，求得y_1，再通过$y-y_1=f^{\prime}(x_1)(x-x_1)$函数找到与x轴相交的点x_2，继而再找到y_2，以此类推，总会找到$f(x)$与x轴相交的点$x_0$

该方法就是牛顿法。

而对于梯度下降法而言，它的公式为：
$$x=x-lr*\nabla f(x)=x-lr*f^{\prime}(x)$$
从而，可以通过控制lr缓慢走到最小值点。lr是学习速率，lr通常不会太大

而如果将SGD换到牛顿法中的化，只需将y轴改为$f^{\prime}(x)$即可，这样，求得的与x轴相交的点也就为极小值点。但是这样的缺陷是：
$$x_1=x_0-f^{\prime}(x)*\frac{1}{f^{\prime \prime}(x)}$$

<span style="color:red">需要求二阶导，这对于深度学习的大量数据而言，每个样本求二阶导，计算量过大，从而出现了SGD来替代牛顿法。</span>

# example

```python
import numpy as np
import matplotlib.pyplot as plt #画图模组
```

```python
def fn(x): #定义函数
    fn= x ** 2 
    return fn
```

```python
def derivate(x): #这里偷懒使用差分的结果定义一阶导数的值
    h=0.001
    derivate=(fn(x+h)-fn(x-h))/(2*h)
    return derivate
```

```python
eta=1 #学习速率=0.001
num_iterate=1000 #重复1k次

x=2
x_value=[x] #把所有的x装进去
fn_value=[fn(x)] #所有目标函数装进去

for i in range(num_iterate):
    x-=eta*(derivate(x))
    x_value.append(x)
    fn_value.append(fn(x))
```

上面x=x-1也可以写成x-=1

```python
x_value #x从2慢慢往0靠近
```

```python
fn_value #也渐渐往0靠近
```

```python
y=np.max(np.abs(x_value))
z=np.linspace(-y,y,num_iterate) #等差数列
```

```python
z
```

   * （一） linspace  函数 是创建等差数列的函数 

*   1.  x = np.linspace(1,10)  #代表 创建1-10的等差数组 ，默认 是50个
*  2.  x = np.linspace(1,10,x)  # x可以设置任意大小

```python
plt.plot(x_value,fn_value,'bo') #右边蓝表示学习了多少内容
plt.plot(z,fn(z),'r') #底层红模型
plt.xlabel('x')
plt.ylabel('function value')
plt.show()
#当学习速率eta达到1时，在两个点反复横跳
```
